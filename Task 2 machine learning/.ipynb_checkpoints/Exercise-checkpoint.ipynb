# Data Preparation

now we loaded and explored our dataset, now time for preparing our data and training a machine learning

for machine learning we usually use python as it has many free useful libraries for machine learning 

what we will be using this for our simple models is week is scikit-learn library

Scikit-learn is a free machine learning library for the Python, It includes various classification, regression and clustering algorithms including Linear Models, Decision Trees, K-Means and K-Nearest Neighbor 

for more information refer to it's guide
https://scikit-learn.org/stable/user_guide.html

so first thing we need to split our data to train, validation and test datasets

Scikit-learn has fucntion that split data easily called train_test_split

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
from sklearn.model_selection import train_test_split


#first split dataset to training set and testing set
train, test = train_test_split(dataset, test_size = 0.2,random_state=43)
print("first split",train.shape,test.shape)

#then split the training set to training set and validation set
train, val = train_test_split(train, test_size = 0.25,random_state=43)
print("second split",train.shape,test.shape,val.shape)

#oringinal dataset is (1797, 65)
```

random_state paramter given to the function is any number we can choose randomly 
as the dataset is splitted randomly according to this number

for more information refer to https://stackoverflow.com/questions/28064634/random-state-pseudo-random-number-in-scikit-learn


we also need to split different dataset to features alone and labels alone using indexing https://numpy.org/doc/stable/user/basics.indexing.html

```python
X_train = train[:,:-1]
Y_train = train[:,-1]
print("train set",X_train.shape,Y_train.shape)

X_test = test[:,:-1]
Y_test = test[:,-1]
print("test set",X_test.shape,Y_test.shape)

X_val = val[:,:-1]
Y_val = val[:,-1]
print("val set",X_val.shape,Y_val.shape)

#X is features and Y is label
```

# Building Machine Learning Model

now time to train our model

in this notebook we will only train a decision tree model

for building decision tree model we will use Scikit-learn

first we will need to refer to the user guide https://scikit-learn.org/stable/user_guide.html

and there we will find the Decision Trees manuel
https://scikit-learn.org/stable/modules/tree.html

```python
#import model library
from sklearn import tree
```

for defining the decision tree we will sklearn.tree.DecisionTreeClassifier
for more information refer to https://scikit-learn.org/stable/modules/tree.html

we will see that this classifier take many parameters what we will focus on only is the max_depth

important note random_state is not a hyper paramter and you should not keep changing it to improve your model

```python
#define model
clf = tree.DecisionTreeClassifier(max_depth=5,random_state=43)
```

```python
#train model
clf = clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_val)
print("predicted labels of the model\n",y_pred)
```

```python
print("ground truth labels\n",Y_val)
```

to check the accuary of the model we can iteratie through the above 2 list and count missclassfied examples
but Scikit-learn offers us easier why for evalution our models
https://scikit-learn.org/stable/modules/model_evaluation.html

we find a find sklearn.metrics.accuracy_score function which we will use https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_val, y_pred)
```

we can see our validtion accuary is not that good

so we need to change our hyperparameter "max depth"

```python
#define model
clf = tree.DecisionTreeClassifier(max_depth=15,random_state=43)
```

```python
clf = clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_val)
```

```python
accuracy_score(Y_val, y_pred)
```

by changing our hyperparamter we got much better accuary

which we may accept depending on our required accaury

if we want our accuary to be above 90% then we will need to keep changing our hyper parameters

but here lets set our acceptance threshold to be  80% and so we can move on to testing phase

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred)
```

our testing accuary passed 80% so this mean we can accept this model and use it in our task and new unseen data

# summary

what we did of loading,preparing, training and evalution can be summarized in the folllowing few lines

```python
#loading data
dataset = np.loadtxt('data.csv')

#preparing data
train, test = train_test_split(dataset, test_size = 0.2,random_state=43)
train, val = train_test_split(train, test_size = 0.25,random_state=43) 

#building and training model
clf = tree.DecisionTreeClassifier(max_depth=15,random_state=43)
clf = clf.fit(train[:,:-1], train[:,-1])

#validation
y_pred = clf.predict(val[:,:-1])
print("validation accuarcy:",accuracy_score(val[:,-1], y_pred))

#testing
y_pred = clf.predict(test[:,:-1])
print("testing accuarcy:",accuracy_score(test[:,-1], y_pred))
```
